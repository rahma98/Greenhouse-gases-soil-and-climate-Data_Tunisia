## Insight:



1. **GWETTOP (MERRA-2 Surface Soil Wetness)** :
   - Un niveau critique pour la surface du sol peut tre considr comme infrieur  environ 20 % d'humidit du sol.
   - Les valeurs infrieures  ce seuil peuvent indiquer un sol sec, peu propice  la croissance des cultures sans irrigation supplmentaire.

2. **GWETROOT (MERRA-2 Root Zone Soil Wetness)** :
   - Pour la zone racinaire du sol, un seuil critique pourrait tre autour de 40 %  50 % d'humidit du sol.
   - Des valeurs infrieures  ce seuil pourraient indiquer un stress hydrique pour les plantes, affectant leur capacit  absorber l'eau ncessaire.

3. **GWETPROF (MERRA-2 Profile Soil Moisture)** :
   - Pour l'humidit sur toute la profondeur du profil du sol, un seuil critique pourrait tre autour de 30 %  40 % d'humidit du sol.
   - Des valeurs infrieures  ce seuil peuvent indiquer un sol excessivement sec, affectant la disponibilit en eau pour les racines oduction agricole.

Les valeurs du GWETPROF (MERRA-2 Profile Soil Moisture) et du GWETROOT (MERRA-2 Root Zone Soil Wetness) se situent entre 28 % et 25 %, cela peut indiquer des conditions de sol relativement sches. 

1. **Conditions de Stress Hydrique** :
   - Des valeurs de GWETPROF et GWETROOT entre 28 % et 25 % suggrent que le sol peut tre en situation de stress hydrique modr  svre.
   - Les plantes peuvent avoir des difficults  accder  suffisamment d'eau pour maintenir leur croissance normale et leur dveloppement racinaire.

2. **Besoin d'Irrigation** :
   - Dans ces conditions, il peut tre ncessaire d'irriguer les cultures pour compenser le manque d'eau dans le sol.
   - L'irrigation peut aider  maintenir des niveaux d'humidit du sol plus favorables pour les plantes et  soutenir la croissance des cultures.

3. **Risque de Dficit Hydrique** :
   - Des valeurs infrieures  30 % dans le GWETPROF et le GWETROOT indiquent un risque accru de dficit hydrique, ce qui peut affecter la productivit des cultures et la qualit des sols.

4. **Impact sur les Rserves d'Eau** :
   - Des valeurs comprises entre 28 % et 25 % dans le GWETPROF et le GWETROOT peuvent indiquer une diminution des rserves d'eau disponibles dans le sol, ce qui peut avoir un impact sur la disponibilit en eau pour les cultures et leses sur le terrain.on agricole.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming 'df' is your DataFrame containing the meteorological features

def plot_category(df, category, variables, ylabel, filter_value=-999):
    """
    Plot variables of a specified category from the DataFrame, excluding negative and specified filter values.
    
    Parameters:
        df (DataFrame): DataFrame containing the meteorological features.
        category (str): Category name for the plot title.
        variables (list): List of variable names to plot from the DataFrame.
        ylabel (str): Label for the y-axis of the plot.
        filter_value (numeric): Value to exclude from plotting (default: -999).
    """
    # Copy the DataFrame to avoid modifying the original
    filtered_df = df.copy()
    
    # Exclude rows with negative or specified filter values in the specified variables
    for variable in variables:
        filtered_df = filtered_df[(filtered_df[variable] > 0) & (filtered_df[variable] != filter_value)]
    
    # Set figure size
    plt.figure(figsize=(13, 9))
    
    # Plot each variable against the 'Date'
    for variable in variables:
        plt.plot(filtered_df['Date'], filtered_df[variable], label=variable)
    
    # Customize plot attributes
    plt.title(f'{category} Over Time of Kairouan')
    plt.xlabel('Date')
    plt.ylabel(ylabel)
    plt.legend()
    plt.grid(True)
    
    # Format x-axis tick labels to display only the years
    plt.gca().xaxis.set_major_formatter(plt.matplotlib.dates.DateFormatter('%Y'))
    plt.gca().xaxis.set_major_locator(plt.matplotlib.dates.YearLocator())
    
    # Rotate x-axis tick labels for better readability (optional)
    plt.xticks(rotation=45)
    
    # Show plot
    plt.show()

# Define categories and corresponding variables
categories = {
    'Radiation Variables': ['ALLSKY_SFC_SW_DWN', 'CLRSKY_SFC_SW_DWN', 'ALLSKY_KT', 
                            'ALLSKY_SFC_LW_DWN', 'ALLSKY_SFC_PAR_TOT', 'CLRSKY_SFC_PAR_TOT',
                            'ALLSKY_SFC_UVA', 'ALLSKY_SFC_UVB', 'ALLSKY_SFC_UV_INDEX'],
    'Wind Speed Variables': ['WS10M', 'WS10M_RANGE', 'WD10M', 'WS50M', 'WS50M_RANGE', 'WD50M', 'WS2M'],
    'Meteorological Variables': ['T2M', 'T2MDEW', 'T2MWET', 'TS', 'T2M_RANGE', 'T2M_MAX', 'T2M_MIN',
                                 'QV2M', 'RH2M', 'PS', 'PRECTOTCORR'],
    'Precipitation Variables': ['PRECTOTCORR'],  
    'Humidity Variables': ['QV2M', 'RH2M'], 
    'Temperature Extremes': ['T2M_MAX', 'T2M_MIN', 'T2M_RANGE'],
    'Surface Pressure': ['PS'],
    'Insolation and UV Index': ['ALLSKY_SFC_UVA', 'ALLSKY_SFC_UVB', 'ALLSKY_SFC_UV_INDEX'],
    'Wind Direction Variables': ['WD10M', 'WD50M']  
}

# Plot each category of meteorological variables with filtering
for category, variables in categories.items():
    plot_category(df, category, variables, ylabel='Value', filter_value=-999)
```

## Air Column Concentration of Ozone

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#read the data 
ozone = pd.read_csv('Monthly Air Column Concentration of Ozone.csv')

# Create DataFrame
ozone = pd.DataFrame(ozone)

# Split the single column into multiple columns
ozone = ozone['DATETIME,TIME,LON,LAT,TOTAL_COLUMN_OZONE'].str.split(',', expand=True)
```

```python
# Generate monthly date range
date_range = pd.date_range(start='2004-10-01', end='2023-08-01', freq='M')

# Create a new DataFrame with monthly dates
df_dates = pd.DataFrame(date_range, columns=['Date'])

# Merge original DataFrame with the date DataFrame based on index
merged_ozone = pd.concat([df_dates, ozone], axis=1)  # Concatenating along columns
```

```python
merged_ozone[4] = merged_ozone[4].astype(float)
```

```python

# Plotting
dates =  merged_ozone['Date']
values =  merged_ozone[4]

plt.figure(figsize=(10, 6))
plt.plot(dates, values, marker='o', linestyle='-', color='b')
plt.xlabel('Date')
plt.ylabel('Value')
plt.title('Monthly Air Column Concentration of Ozone over Kairouan')
plt.xticks(rotation=90)
plt.grid(True)
plt.tight_layout()
plt.show()
```

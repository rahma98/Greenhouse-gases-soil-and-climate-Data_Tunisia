# I. Famine Early Warning Systems Network Land Data Assimilation System Surface Soil Moisture Anomalies

```python
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('FLDAS Surface Soil Moisture Anomalies.csv')
```

```python
data.info()
```

```python
data['date']= pd.to_datetime(data['date'])
```

```python
data.head(5)
```

```python
df = pd.DataFrame(data)

# Specify columns for the bar chart
columns_to_plot = ['min', 'max', 'mean', 'max', 'std','median']

# Subset the DataFrame to include only the specified columns
subset_df = df[columns_to_plot]

```

```python
# Plotting the line graph
plt.figure(figsize=(12, 8))
for column in columns_to_plot:
    plt.plot(df['date'], df[column], label=column)

plt.title('FLDAS Surface Soil Moisture Anomalies in Tunisia')
plt.xlabel('Date')
plt.ylabel('values')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.legend(title='Columns', bbox_to_anchor=(1.05, 1), loc='upper left')  # Add legend with column names as titles
plt.grid(True, linestyle='--', alpha=0.7)  # Add grid lines
plt.show()
```

## Insight:



1. **GWETTOP (MERRA-2 Surface Soil Wetness)** :
   - Un niveau critique pour la surface du sol peut tre considr comme infrieur  environ 20 % d'humidit du sol.
   - Les valeurs infrieures  ce seuil peuvent indiquer un sol sec, peu propice  la croissance des cultures sans irrigation supplmentaire.

2. **GWETROOT (MERRA-2 Root Zone Soil Wetness)** :
   - Pour la zone racinaire du sol, un seuil critique pourrait tre autour de 40 %  50 % d'humidit du sol.
   - Des valeurs infrieures  ce seuil pourraient indiquer un stress hydrique pour les plantes, affectant leur capacit  absorber l'eau ncessaire.

3. **GWETPROF (MERRA-2 Profile Soil Moisture)** :
   - Pour l'humidit sur toute la profondeur du profil du sol, un seuil critique pourrait tre autour de 30 %  40 % d'humidit du sol.
   - Des valeurs infrieures  ce seuil peuvent indiquer un sol excessivement sec, affectant la disponibilit en eau pour les racines oduction agricole.

Les valeurs du GWETPROF (MERRA-2 Profile Soil Moisture) et du GWETROOT (MERRA-2 Root Zone Soil Wetness) se situent entre 28 % et 25 %, cela peut indiquer des conditions de sol relativement sches. 

1. **Conditions de Stress Hydrique** :
   - Des valeurs de GWETPROF et GWETROOT entre 28 % et 25 % suggrent que le sol peut tre en situation de stress hydrique modr  svre.
   - Les plantes peuvent avoir des difficults  accder  suffisamment d'eau pour maintenir leur croissance normale et leur dveloppement racinaire.

2. **Besoin d'Irrigation** :
   - Dans ces conditions, il peut tre ncessaire d'irriguer les cultures pour compenser le manque d'eau dans le sol.
   - L'irrigation peut aider  maintenir des niveaux d'humidit du sol plus favorables pour les plantes et  soutenir la croissance des cultures.

3. **Risque de Dficit Hydrique** :
   - Des valeurs infrieures  30 % dans le GWETPROF et le GWETROOT indiquent un risque accru de dficit hydrique, ce qui peut affecter la productivit des cultures et la qualit des sols.

4. **Impact sur les Rserves d'Eau** :
   - Des valeurs comprises entre 28 % et 25 % dans le GWETPROF et le GWETROOT peuvent indiquer une diminution des rserves d'eau disponibles dans le sol, ce qui peut avoir un impact sur la disponibilit en eau pour les cultures et leses sur le terrain.on agricole.

```python
# Create an empty dictionary to store counts for each column
counts_999 = {}

# Iterate over each column in the DataFrame
for column in merged_df.columns:
    # Count occurrences of -999 in the current column
    count_999 = (merged_df[column] == -999).sum()
    
    # Store the count in the dictionary with column name as the key
    counts_999[column] = count_999

# Display the counts of - 999 for each column
for column, count in counts_999.items():
    print(f"Column '{column}': {count} occurrences of -999")

# there an important null values in the columns CLRSKY_SFC_PAR_TOT ,  ALLSKY_SFC_UVA and ALLSKY_SFC_UVB !!!! 
```

```python
# Create a new column 'Date' by merging 'YEAR', 'MO', and 'DY' with '/' separator
df['Date'] = df['YEAR'].astype(str) + '/' + df['MO'].astype(str) + '/' + df['DY'].astype(str)

# Convert the concatenated date strings to datetime format
df['Date'] = pd.to_datetime(df['Date'], format='%Y/%m/%d')
```

```python
# List of original date columns to be removed
date_columns = ['YEAR', 'MO', 'DY']

# Drop the original date columns from the DataFrame
df.drop(date_columns, axis=1, inplace=True)

# save the data
df.to_csv('climate.csv', index=False)
```
